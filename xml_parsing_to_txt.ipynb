```python
from xml.etree.ElementTree import parse
import os,glob
```

```python
path = 'C:/Users/User/Downloads/archive/annotations_jpg'
```

```python
def sorted_list(path):
    tmplist = glob.glob(path)
    tmplist.sort()
    
    return tmplist
```

```python
list_data = sorted_list(os.path.join(path,'*.xml'))
```

```python
list_data
```

```python
for i in list_data[:10]:
    file=i.split('/')[-1].split('.')[0].split('\\')[-1]
    tree=parse(i)
    root = tree.getroot()
    face = root.findall('object')
    name = [x.findtext('name') for x in face]
    xmin = []
    ymin = []
    xmax = []
    ymax = []
    
    for object in root.iter('object'):
        xmin.append(object.find('bndbox').findtext('xmin'))
        ymin.append(object.find('bndbox').findtext('ymin'))
        xmax.append(object.find('bndbox').findtext('xmax'))
        ymax.append(object.find('bndbox').findtext('ymax'))
        
    f = open(path+f'{file}.txt','w')
    
    for j in range(len(xmin)):
        if name[j] != 'face':
            name[j] = 'face'
        data = '%s %s %s %s %s\n'%(name[j],xmin[j],ymin[j],xmax[j],ymax[j])
        
        f.write(data)
    f.close()     
```
